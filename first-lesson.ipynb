```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
```

```python
samples = 100

# Step 1: Generate synthetic data
np.random.seed(0)
sizes = np.random.normal(1000, 300, samples)
prices = 200000 + (sizes * 100) + np.random.normal(0, 10000, samples)
plt.scatter(sizes, prices, color='blue')
plt.xlabel('Size')
plt.ylabel('Price')
plt.title('Home size vs Price')
plt.show()
sizes = np.array(sizes).reshape(-1, 1)
prices = np.array(prices).reshape(-1, 1)
```

```python
# Train a model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(sizes, prices)

print('Intercept:', int(model.intercept_))
print('Slope:', int(model.coef_))
```

```python
# Visualize the test data and the regression line
plt.scatter(sizes, prices, color='blue', label='Data')
plt.plot(sizes, model.predict(sizes), color='green', linewidth=2, label='Regression Line')
plt.xlabel('Size')
plt.ylabel('Salary')
plt.title('Linear Regression: Size vs Salary')
plt.legend()
plt.show()
```
